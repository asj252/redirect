## Inner product of Vectors

The dot product is a fundamental concept in linear algebra, and the following two definitions are equivalent for finite-dimensional vector spaces.

Let $\mathbf{u}=\boldsymbol{\vec u}=[u_1,u_2,...u_n]$ and $\mathbf{v}=[v_1,v_2,...v_n]$ be two vectors of the same dimension. The dot product is defined as:
$$\langle \mathbf{u},\mathbf{v}\rangle=\sum_{i=1}^{i=n} u_i v_i$$

Alternatively, it can be defined as:
$$\langle \mathbf{u},\mathbf{v} \rangle=|\mathbf{u}||\mathbf{v}|cos\theta$$

Here, $\theta$ represents the angle between vectors $\mathbf{u}$ and $\mathbf{v}$ in the Euclidean plane spanned by them.$|\mathbf{u}|$ means the Euclidean distance of $\mathbf{u}$.

## Cauchy-Schewarz inequality

From the second definition of the dot product, we can derive the Cauchy-Schwarz inequality:

$$\sum_{i=1}^{n} u_i v_i=\langle \boldsymbol{\vec u},\boldsymbol{\vec v}\rangle \leq |\boldsymbol{\vec u}||\boldsymbol{\vec v}|=\sqrt{\sum_{i=1}^{n} u_i^2 \sum_{i=1}^{n} v_i^2}$$

From the Cauchy-Schwarz inequality(Why? Math challenge), we can also obtain the triangle inequality:

$$| \boldsymbol{\vec u}+\boldsymbol{\vec v} |\leq|\boldsymbol{\vec u}|+|\boldsymbol{\vec v}|$$

If $|\boldsymbol{\vec v}|=1$, meaning it is a unit vector, then $\langle \boldsymbol{\vec u},\boldsymbol{\vec v}\rangle \boldsymbol{\vec v} = \boldsymbol{\vec u}|_\vec v$ can be seen as the component of $\boldsymbol{\vec u}$ in the direction of $\boldsymbol{\vec v}$.

```python
u = np.array([1,2,3])
v = np.array([4,5,6])

np.dot(u,v)
```

## Matrix
In linear algebra, a matrix is an important mathematical tool that can represent a rectangular arrangement of numbers. Matrices can be used to represent linear transformations and are crucial for solving problems related to linear equations and vector spaces.

A matrix is composed of one or more rows and columns. Each row represents a vector, and each column represents a component. For example, a 2-row by 3-column matrix can be represented as:

$$\mathbf{M}=\begin{bmatrix}1 & 2 & 3 \\4 & 5 & 6 \end{bmatrix}$$

```python
M = np.array([[1,2,3],[4,5,6]])
```

The application of matrices in linear algebra is very extensive. For example, matrices can be used to represent linear transformations such as rotation, translation, scaling, mirroring, and projection. Furthermore, matrices can be employed to solve systems of linear equations by combining the coefficient and constant matrices into an augmented matrix and solving it using methods like Gaussian elimination or LU decomposition.

In addition, matrices can be used to represent linear transformations in vector spaces and to calculate operations such as projections in vector spaces. In computer science, matrices are widely applied in areas such as image processing, machine learning, and data analysis.

In summary, matrices are a fundamental concept in linear algebra, providing us with powerful tools to understand linear transformations, solve systems of linear equations, and perform various operations in vector spaces.

## Change Bases
We can view matrix as a way to do basis change
What is a basis?
We call $\mathbf{e}_1=[1,0]$, and $\mathbf{e}_2=[0,1]$ as the standard basis for $\mathbf{R}^2$: Every vector in $\mathbf{R}^2$ can be represented as an unique combination of $\mathbf{e}_1$ and $\mathbf{e}_2$. *Can you prove the statement?* 

Other than the standard basis, there infinite many bases for $\mathbf{R}^2$.

```python
#For example the vector [1,2],[1,4] can also form a basis for R^2
# Vector origin location
X0_i = [0,0]
Y0_i = [0,0]
  
# Directional vectors
X1_i = [1,1]  
Y1_i = [2,4]  
  
# Creating plot
plt.quiver(X0_i, Y0_i, X1_i, Y1_i, color='b', units='xy', scale=1)
plt.title('Another basis')
  
# x-lim and y-lim
plt.xlim(-1, max(X1_i)+1)
plt.ylim(-1, max(Y1_i)+1)

#get current axes and set the aspect ratio
plt.gca().set_aspect('equal')

# Show plot with grid
plt.grid()
plt.show()
```

As long as any two vectors are not pararrel, they can form a basis for $\mathbf{R}^2$.
For example: $\mathbf{u}_1=[1,2]$, and $\mathbf{u}_2=[1,4]$ could form a basis. 
If $\mathbf{v}=x\mathbf{u}_1 +y\mathbf{u}_2 $, therefore $\mathbf{v} =[x,y]_{\mathbf{u}_1,\mathbf{u}_2}$ in the $\mathbf{u}_1, \mathbf{u}_2 $ reference coordinate. It is easy to calculate $\mathbf{v}$ in terms of the standard basis.
$\mathbf{v}=[x\times 1+y\times1,x\times 2+y\times 4]_{\mathbf{e}_1,\mathbf{e}_2}=[x+y,2x+4y]$. 
We mostly skip the subscripts for standard basis.

We could also represent this in terms of matrix

$$\mathbf{v}=\begin{bmatrix}\mathbf{u}_1 & \mathbf{u}_2 \end{bmatrix}\begin{bmatrix} x \\ y \end{bmatrix}_{\mathbf{u}_1,\mathbf{u}_2}=\begin{bmatrix}1 & 1 \\2 &4 \end{bmatrix}\begin{bmatrix} x \\ y \end{bmatrix}=\begin{bmatrix} x+y \\ 2x+4y \end{bmatrix}_{\mathbf{e}_1,\mathbf{e}_2}$$

## Chicken and Rabbit problem
Suppose we have and only have a bunch of normal chicken and normal rabbits. We could use the number of chicken $x$ and the number of rabbit $y$ to represent current state. On the other hand we could use the number of  heads $a$ and the number of legs $b$ to represent the current state.  

Obvisouly this two representations are linked together as 
$$\begin{bmatrix} a \\ b \end{bmatrix}_{head,leg}=\begin{bmatrix}1 & 1 \\2 &4 \end{bmatrix}\begin{bmatrix} x \\ y \end{bmatrix}_{chicken,rabbit}=\begin{bmatrix} x+y \\ 2x+4y \end{bmatrix}_{head,leg}$$

Or we write it down in a more familiar system of linear equations form
$$
\left\{
\begin{align*}
x + y &= a \\
2x + 4y &= b \\
\end{align*}
\right.
$$

It is easy to see every matrix is equivalent to a system of linear equations.

## Inverse Matrix
We can write down linear equation in the form of 
$$\mathbf{M}\mathbf{x}=\mathbf{b}$$

where $\mathbf{M}$ is a matrix, $\mathbf{x}$,$\mathbf{b}$ are vectors.


For square matrix, we could consider its inverse $\mathbf{M}^{-1}$.

For a 2 by 2 matrix $\mathbf{M} =\begin{bmatrix}a & b \\c &d \end{bmatrix}$, we even have a formula:

$$\mathbf{M}^{-1} =\frac{1}{det(M)}\begin{bmatrix}d & -b \\-c &a \end{bmatrix}$$

let's check it

```python
M = sp.Matrix([[a,b],[c,d]])
Minv = sp.Matrix([[d,-b],[-c,a]])
M*Minv
```

Therefore, for 2 by 2 matrix, if and only if $det(M)\neq 0$, then $\mathbf{M}\mathbf{x}=\mathbf{b}$ have an unique solution.
